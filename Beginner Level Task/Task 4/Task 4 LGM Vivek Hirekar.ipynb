## Task 2:- Image to Pencil sketch in Python

```python
import cv2
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('dog.jpg')
```

```python
def show_image(img):
    plt.figure(dpi=150)
    rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.axis('Off')
    plt.show()
```

```python
show_image(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap=mpl.cm.binary_r);
plt.axis('off');
```

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.imshow(inverted_image, cmap=mpl.cm.binary_r);
plt.axis('off');
```

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),0)
plt.imshow(img_smoothing, cmap=mpl.cm.binary_r);
plt.axis('off');
```

```python
inverted_blurred = 255 - img_smoothing
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.imshow(pencil_sketch, cmap=mpl.cm.binary_r);
plt.axis('off');
```
